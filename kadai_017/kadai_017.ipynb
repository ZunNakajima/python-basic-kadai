```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f'{self.name}さんは大人です。')
    else:
      print(f'{self.name}さんは大人ではないです')

person_list = []
person_list.append(Human('太郎', 15))
person_list.append(Human('花子', 25))
person_list.append(Human('健太', 19))
person_list.append(Human('淳', 20))
person_list.append(Human('純一郎', 21))

for person in person_list:
  person.check_adult()
```
